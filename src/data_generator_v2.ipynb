```python
## Generating random values with timestamp

# Import libraries

import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Generating random temperature values with timestamp

with open(f"T&P_dataset_v2.csv", "w") as file:
    file.write("timestamp,temperature,pressure\n")

    current_time = datetime(2026, 6, 11, 6, 00) 
    base_temperature = 40
    temperature=[]
    timestamp=[]
    
    for i in range(3):
        for j in range(20):
            current_time += timedelta(minutes=1)            
            increment = random.randint(-2, 2)
            base_temperature += increment
            timestamp.append(current_time)
            temperature.append(base_temperature)
                        
        for k in range(10):
            current_time += timedelta(minutes=1) 
            increment_1 = random.randint(2, 8)
            base_temperature += increment_1
            timestamp.append(current_time)
            temperature.append(base_temperature)
            
        for l in range(3):
            current_time += timedelta(minutes=1)
            increment_2 = random.randint(20, 30)
            base_temperature += increment_2
            timestamp.append(current_time)
            temperature.append(base_temperature)
            
        for m in range(8):
            current_time += timedelta(minutes=1)
            increment_3 = random.randint(11, 21)
            base_temperature -= increment_3
            timestamp.append(current_time)
            temperature.append(base_temperature)
            
        for n in range(20):
            current_time += timedelta(minutes=1)
            increment_4 = random.randint(-2, 2)
            base_temperature += increment_4
            timestamp.append(current_time)
            temperature.append(base_temperature)
            
    print(len(temperature))
    print(temperature)
    print(len(timestamp))
    print(type(timestamp[0]))
```

```python
# Generating random pressure values

base_pressure = 1.5
pressure=[]
s=0
length=0
length=len(temperature)-10

for o in range(1):
        for p in range(10):
            increment_5 = random.uniform(-0.1,0.1)
            base_pressure += increment_5
            pressure.append(round(base_pressure,2))
        for q in range (length):
            increment_6 = random.uniform(0,0.5)
            increment_7 = random.uniform(0,0.62)
            if temperature[s+1] > temperature[s]:
                base_pressure += increment_6
                pressure.append(round(base_pressure,2))
                s += 1
            elif temperature[s+1] < temperature[s]:
                base_pressure -= increment_7
                pressure.append(round(base_pressure,2))
                s += 1
            else:
                s += 1
                pressure.append(round(base_pressure,2))
                
                    
                
print(pressure)
print(len(pressure))
```

```python
# Visualization

time = timestamp
temp = temperature
press = pressure

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(time, temp, color=color, label="Temperature")
ax1.tick_params(axis='y', labelcolor=color)
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b-%d %H:%M'))
fig.autofmt_xdate()


ax2 = ax1.twinx() 


color = 'tab:blue'
ax2.set_ylabel('Pressure (bar)', color=color)  
ax2.plot(time, press, color=color, label="Pressure")
ax2.tick_params(axis='y', labelcolor=color)


lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(
    lines1 + lines2,
    labels1 + labels2,
    bbox_to_anchor=(0.5, -0.4),
    loc='upper center',
    ncol=2
)

plt.title('Temperature and Pressure Monitoring Dashboard')

fig.tight_layout()  
plt.show()
```

```python
# Statistical alerts

temperature_rising_limit_s = df_process_data["temperature"].quantile(0.85)
temperature_critical_limit_s = df_process_data["temperature"].quantile(0.95)
pressure_rising_limit_s = df_process_data["pressure"].quantile(0.85)
pressure_critical_limit_s = df_process_data["pressure"].quantile(0.95)
s=0

df_statistical_alerts = pd.DataFrame(columns=["timestamp", "variable", "level", "value"])

for index, row in df_process_data.iterrows():
    if row["temperature"] > temperature_rising_limit_s and row["temperature"] < temperature_critical_limit_s :
        df_statistical_alerts.loc[s]=[row["timestamp"], "temperature", "warning", row["temperature"]]
        s += 1
        
    if row["temperature"] > temperature_critical_limit_s:
        df_statistical_alerts.loc[s]=[row["timestamp"], "temperature", "critical", row["temperature"]]
        s += 1

    if row["pressure"] > pressure_rising_limit_s and row["pressure"] <pressure_critical_limit_s:
        df_statistical_alerts.loc[s]=[row["timestamp"], "pressure", "warning", row["pressure"]]
        s += 1

    if row["pressure"] > pressure_critical_limit_s:
        df_statistical_alerts.loc[s]=[row["timestamp"], "pressure", "critical", row["pressure"]]
        s += 1
        
print(df_statistical_alerts)
```
